```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
df=pd.read_csv("Automobile_data.csv")
```

```python
sb.pairplot(df)
```

```python
#deal with missing data in price column
df['price']=df['price'].fillna(method='ffill')
```

# count total cars per company

```python
c = df['company'].unique()
n = []
for i in c:
    n.append(df['company'].loc[df['company']==i].count())
fif=plt.figure(figsize=(15,8))
ax=fif.add_axes([0,0,1,1])
ax.bar(c,n,align='center',width=0.5)
plt.show()
```

```python
df1=pd.Series(n)
df2=pd.Series(c)
sb.set(rc={"figure.figsize":(15,10)})
sb.barplot(x=df2,y=df1)
```

# Find each companys Highest price car

```python
g=df.groupby(by="company")
print(g['price'].max())
for name,group in g:
    print(name,end='  ')
    print(group['price'].max(),'\n')
```

```python
x=[]
y=[]
g=df.groupby(by="company")
for name,group in g:
    x.append(name)
    y.append(group['price'].max())
fif=plt.figure(figsize=(15,8))
ax=fif.add_axes([0,0,1,1])
ax.bar(x,y,align='center',width=0.5)
plt.xlabel("car company")
plt.ylabel("max price")
plt.show()
```

# Find the average mileage of each car making company

```python
g=df.groupby(by="company")
x=[]
y=[]
for name,group in g:
    x.append(name)
    y.append(group['average-mileage'].agg(np.mean))
fif=plt.figure(figsize=(15,8))
ax=fif.add_axes([0,0,1,1])
ax.bar(x,y,align='center',width=0.5)
plt.xlabel("car company")
plt.ylabel("average mileage")
plt.show()
```
