```python
n1=df["type"].loc[df["type"]=="Movie"].count()#Q1
n2=df["type"].loc[df["type"]=="TV Show"].count()
print("number of movie and tv shows are",n1,"and",n2)
n=n1+n2
print("Percent of tv shows is",round(n2/n*100,2))
print("Percent of movies is",round(n1/n*100,2))
```

```python
g=df.groupby("release_year")#Q2 and Q3....Q4 can also explained from this output
for i,group in g:#Here i gives year value and group gives whol
    m=g.get_group(i)
    n1=m["type"].loc[m["type"]=="Movie"].count()
    n2=m["type"].loc[m["type"]=="TV Show"].count()
    print("Count of Movie and tv shows in year",i,"is",n1,"and",n2)
```

```python
print(df['duration'])#Q5
g=df.groupby(by="type")
m=g.get_group("Movie")
t=g.get_group("TV Show")
a=m.loc[:,'duration'].astype(str)
a=a.str.slice(0,-3,1)
a=a.astype(int)
print("Average of movies time are",a.mean())
```

```python
t=g.get_group("TV Show")#Q6
b=t.loc[:,'duration'].astype(str)
b=b.str.slice(0,-7)
#b=b.str.slice(0,2)
b=b.astype(int)
print("Average of tv shows seasons are",b.mean())
```

```python
g=df.groupby(by="type")#Q7 and Q8
m=g.get_group("Movie")
t=g.get_group("TV Show")
n=5
print("Top 5 movie directors are \n",m['director'].value_counts()[:n].index.tolist())#####Here if we dont use index it gives us the num of counts only
print("Top 5 tv shows directors are \n",t['director'].value_counts()[:n].index.tolist())
```

```python
g=df.groupby(by="type")#Q9
m=g.get_group("Movie")
t=g.get_group("TV Show")
n=3
print("Top 3 countries with most movies are \n",m['country'].value_counts()[:n].index.tolist())
print("Top 3 countries with most tv shows are \n",t['country'].value_counts()[:n].index.tolist())
```

```python
#Q10 and 11
g1=df.groupby("release_year")
g=df.groupby(by=['type','country'])
a=g.get_group(("TV Show","United States"))
b=g.get_group(("Movie","United States"))
for i,group in g1:
    Movies_count=a['release_year'].loc[a['release_year']==i].count()
    TV_shows_count=b['release_year'].loc[b['release_year']==i].count()
    print("Movie in year",i," in US is",Movies_count)
    print("Tv shows in year",i,"in US is",TV_shows_count)
```

```python
g=df.groupby(by="type")#Q12 and Q13
m=g.get_group("Movie")
t=g.get_group("TV Show")
print("Most famous cast in movies are \n",m['cast'].mode().tolist())
print("Most famous cast in tv shows are \n",t['cast'].mode().tolist())
```
