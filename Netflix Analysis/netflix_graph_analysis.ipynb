```python
n1=df["type"].loc[df["type"]=="Movie"].count()#Q1
n2=df["type"].loc[df["type"]=="TV Show"].count()
t=[n1,n2]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(t,labels=['Movie','TV Shows'],autopct='%1.2f%%')
plt.show()
```

```python
g=df.groupby("release_year")#Q2 and Q3....Q4 can also explained from this output
y1=[]
y2=[]
x1=[]
for i,group in g:#Here i gives year value and group gives whol
    m=g.get_group(i)
    y1.append(m["type"].loc[m["type"]=="Movie"].count())
    y2.append(m["type"].loc[m["type"]=="TV Show"].count())
    x1.append(i)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x1,y1,'y')
l2 = ax.plot(x1,y2,'g') 
ax.legend(labels = ('movie', 'tv shows'), loc = 'upper left') # legend placed at upper lefy
ax.set_xlabel('release year')
ax.set_ylabel('num of movie/tv shows')
plt.show()
```

```python
Movies_count=[]
TV_shows_count=[]
x=[]
g1=df.groupby("release_year")
g=df.groupby(by=['type','country'])
a=g.get_group(("TV Show","United States"))
b=g.get_group(("Movie","United States"))
for i,group in g1:
    Movies_count.append(a['release_year'].loc[a['release_year']==i].count())
    TV_shows_count.append(b['release_year'].loc[b['release_year']==i].count())
    x.append(i)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x,Movies_count,'r') # solid line with yellow colour and square marker
l2 = ax.plot(x,TV_shows_count,'g') # dash line with green colour and circle marker
ax.legend(labels = ('movie', 'tv shows'), loc = 'upper left') # legend placed at lower right
ax.set_xlabel('release year')
ax.set_ylabel('expansion of movie/tv shows in us')
plt.show()
```

```python
fif=plt.figure(figsize=(15,8))
ax=fif.add_axes([0,0,1,1])
ax.bar(x,Movies_count,align='center',width=0.5)
plt.xlabel("release year")
plt.ylabel("expansion of movie in us")
plt.show()
```
