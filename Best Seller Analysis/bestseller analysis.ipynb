```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
df=pd.read_csv("bestsellers with categories.csv")
```

# plot userrating histograph

```python
print(df['User Rating'].unique())
```

```python
a=df['User Rating'].unique()
a.sort()
r=[]
for i in a:
    r.append(df['User Rating'].loc[df['User Rating']==i].count())
b=[]
t=3.3
while(t<=4.9):
    b.append(t)
    t=t+0.1
print(r)
fig,ax = plt.subplots(1,1)
x = np.array(a)
ax.hist(x, bins = b,weights=r)
ax.set_title("histogram of result")
#ax.set_xticks([0,25,50,75,100])
ax.set_xlabel('Ratings')
ax.set_ylabel('Count')
plt.show()
```

# popularity of books

```python
popularity = [0,0,0,0]
for i in df["User Rating"]:
    if i >= 4.8:
        popularity[0] += 1
    elif i >= 4.5:
        popularity[1] += 1
    elif i >= 4.0:
        popularity[2] += 1
    else:
        popularity[3] += 1

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ["Very much popular","very popular","Fairly popular","popular"]
ax.bar(langs,popularity,width=0.6,bottom=5)
plt.show()
```
