# Dealing with missing values/preprocessiing

```python
print(df['Cabin'].unique())
df=df.drop(columns='Cabin',axis=1)
df.head()
```

```python
print(df['Embarked'].unique())
df['Embarked']=df['Embarked'].fillna(method='ffill')
df['Embarked'].unique()
```

```python
g=df.groupby(by='Pclass')
print(g)
print(g['Age'].agg(np.mean),df['Age'].mean())
l = [38,29,25]
i=0
for n,v in g:
    df.loc[df['Pclass']==n,'Age'] = df.loc[df['Pclass']==n, 'Age'].fillna(l[i])
    i+=1
#df.loc[df['Pclass']==1,'Age'] = df.loc[df['Pclass']==1,'Age'].fillna(38)
#df.loc[df['Pclass']==2,'Age'] = df.loc[df['Pclass']==2,'Age'].fillna(29)
#df.loc[df['Pclass']==3,'Age'] = df.loc[df['Pclass']==3,'Age'].fillna(25)
```

# #find out the survival rate

```python
df1=df.copy()
```

```python
n=df1["Survived"].loc[df1["Survived"]==1].count()
print(n)
t=df1["Survived"].count()
rate=n/t*100
rate=round(rate,2)
print("Survival rate is",rate,"%")
```

find out the ratio of gender

```python
r=df1["Sex"].loc[df1["Sex"]=="male"].count()
s=df1["Sex"].loc[df1["Sex"]=="female"].count()
print("Ratio is",round(r/s,2),":1")

t=r+s
rate_male=round(r/t*100,2)
rate_female=round(s/t*100,2)
print("Rate of male and females are",rate_male,"% and",rate_female,"%")
```

# find out the survival rate according to gender

```python
g=df.groupby(by="Sex")
m=g.get_group('male')
f=g.get_group('female')
c=m['Survived'].loc[m['Survived']==1].count()
t=m['Survived'].count()
a=f['Survived'].loc[f['Survived']==1].count()
b=f['Survived'].count()
print("count of male survived",c)
print("count of male total",t)
print("Survival rate of male",round(c/t*100,2),"%")
print("count of female survived",a)
print("count of female total",b)
print("Survival rate of female",round(a/b*100,2),"%")
```

# find the ratio of classes

```python
n1=df1["Pclass"].loc[df1["Pclass"]==1].count()
n2=df1["Pclass"].loc[df1["Pclass"]==2].count()
n3=df1["Pclass"].loc[df1["Pclass"]==3].count()
n=n1+n2+n3
print("Percentage of class 1 is",round(n1/n*100,2),"%")
print("Percentage of class 2 is",round(n2/n*100,2),"%")
print("Percentage of class 3 is",round(n3/n*100,2),"%")
```

# find the survival rate on the basis of classes

```python
g=df.groupby(by="Pclass")
class1=g.get_group(1)
class2=g.get_group(2)
class3=g.get_group(3)
c1_sur=class1['Survived'].loc[class1['Survived']==1].count()
c2_sur=class2['Survived'].loc[class2['Survived']==1].count()
c3_sur=class3['Survived'].loc[class3['Survived']==1].count()
c1_tot=class1['Survived'].count()
c2_tot=class2['Survived'].count()
c3_tot=class3['Survived'].count()
print("Survival rate of class 1 is",round(c1_sur/c1_tot*100,2),"%")
print("Survival rate of class 2 is",round(c2_sur/c2_tot*100,2),"%")
print("Survival rate of class 3 is",round(c3_sur/c3_tot*100,2),"%")
```

# Find out the survival rate of male and female according to class

```python
g=df.groupby(by=['Pclass','Sex'])
print(g.groups)
c1_female=g.get_group((1,"female"))
c1_male=g.get_group((1,"male"))
c2_female=g.get_group((2,"female"))
c2_male=g.get_group((2,"male"))
c3_female=g.get_group((3,"female"))
c3_male=g.get_group((3,"male"))
c1_female_sur=c1_female['Survived'].loc[c1_female['Survived']==1].count()
#print('\n',c1_female_sur)
c1_male_sur=c1_male['Survived'].loc[c1_male['Survived']==1].count()
c2_female_sur=c2_female['Survived'].loc[c2_female['Survived']==1].count()
c2_male_sur=c2_male['Survived'].loc[c2_male['Survived']==1].count()
c3_female_sur=c3_female['Survived'].loc[c3_female['Survived']==1].count()
c3_male_sur=c3_male['Survived'].loc[c3_male['Survived']==1].count()

c1_total=c1_male_sur+c1_female_sur
c2_total=c2_male_sur+c2_female_sur
c3_total=c3_male_sur+c3_female_sur

print('\n')

print("survival rate of female according to class 1 is ",round(c1_female_sur/c1_total*100,2),"%")
print("survival rate of male according to class 1 is ",round(c1_male_sur/c1_total*100,2),"%")
print("survival rate of female according to class 2 is ",round(c2_female_sur/c2_total*100,2),"%")
print("survival rate of male according to class 2 is ",round(c2_male_sur/c2_total*100,2),"%")
print("survival rate of female according to class 3 is ",round(c3_female_sur/c3_total*100,2),"%")
print("survival rate of male according to class 3 is ",round(c3_male_sur/c3_total*100,2),"%")
```
